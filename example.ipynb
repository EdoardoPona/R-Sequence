## Speed

```python

def phi(d): 
    x=2.0000 
    for i in range(10): 
        x = pow(1+x,1/(d+1)) 
    return x



## very basic port of the cpp implementation in numpy 
def quasirandom(d, N, seed=0.5):
    g=2.0000 
    for i in range(10): 
        g = np.power(1+g,1/(d+1)) 
    a = 1.0/np.power(g, np.arange(1, d+1))
    X = []
    current = 0
    for n in range(1, N+1):
        current += a
        X.append((seed + current) % 1)
    return np.vstack(X)
```
